```python
# Mean Median Mode & Std.Dev of Monthly Charges
print("Mean of Monthly Charges: ",df.MonthlyCharges.mean())
print("Median of Monthly Charges: ",df.MonthlyCharges.median())
print("Mode of Monthly Charges: ",df.MonthlyCharges.mode()[0])
print("Standard Deviation of Monthly Charges: ",df.MonthlyCharges.std())
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ', 0)
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
# Mean Median Mode & Std.Dev of Total Charges
print("Mean of Total Charges: ",df.TotalCharges.mean())
print("Median of Total Charges: ",df.TotalCharges.median())
print("Mode of Total Charges: ",df.TotalCharges.mode()[0])
print("Standard Deviation of Total Charges: ",df.TotalCharges.std())
```

```python
# Churn
df.Churn.value_counts()
```

---------------

<b>Some Basic Data Visualizations</b>

```python
# Scatter Plot Between Monthly & Total Charges
plt.figure(figsize=(10,7))
plt.scatter(df.TotalCharges,df.MonthlyCharges)
plt.title("Scatter Plot Between Monthly & Total Charges",fontsize=16)
plt.xticks(np.arange(0,8001,1000))
plt.xlabel("Total Charges",fontsize=13)
plt.ylabel("Monthly Charges",fontsize=13)
plt.show()
```

```python
# Correlation Matrix Of Monthly & Total Charges 

m_t = {"MonthlyCharges": df.MonthlyCharges, "TotalCharges": df.TotalCharges}
charges = pd.DataFrame(data = m_t)

m_t_corr = charges.corr()
plt.figure(figsize= (5,4))
sns.heatmap(m_t_corr, cmap= 'summer', annot= True, linewidths= 1)
plt.suptitle('Correlation Matrix', fontsize= 14)
plt.show()
```

```python
# Overall Correlation Matrix Of Original Data

corr= df.corr()

plt.figure(figsize= (16,10))
sns.heatmap(corr, cmap= 'YlOrRd', annot= True, linewidths= 1)
plt.title('Correlation Matrix', fontsize= 16)
```

-----------

<b>The Gender Analysis & Visualizations</b>

```python
# Gender Counts in the Dataset


plt.figure(figsize=(7,6))
sns.countplot(df.gender)
plt.title("TOTAL COUNTS OF GENDERS")
plt.show()
```

```python
# Genders and their Churn Counts

yes_churn = df[df["Churn"]=="Yes"]
print("==GENDERS HAVING YES CHURN==")
print(yes_churn.gender.value_counts())

no_churn = df[df["Churn"]=="No"]
print("\n==GENDERS HAVING NO CHURN==")
print(no_churn.gender.value_counts())

fig= plt.figure(figsize= (16,5))


fig.add_subplot(1,3,1)
sns.countplot(df.gender, hue= df.Churn)
plt.yticks(np.arange(0,2501,250))
plt.title("BAR PLOT OF GENDERS AND CHURN")

fig.add_subplot(1,3,2)
plt.pie(yes_churn.gender.value_counts(), labels= yes_churn.gender.value_counts().index, autopct= '%1.2f%%', startangle= 90)
plt.title('YES CHURN GENDERS', fontsize= 13)


fig.add_subplot(1,3,3)
plt.pie(no_churn.gender.value_counts(), labels= no_churn.gender.value_counts().index, autopct= '%1.2f%%', startangle= 0)
plt.title('NO CHURN GENDERS', fontsize= 13)


plt.show()
```

```python
# Genders and their Seniority

seniority0 = df[df["SeniorCitizen"]==0]
print("==GENDERS WHO ARE JUNIOR==")
print(seniority0.gender.value_counts())

seniority1 = df[df["SeniorCitizen"]==1]
print("\n==GENDERS WHO ARE SENIOR==")
print(seniority1.gender.value_counts())

fig= plt.figure(figsize= (16,5))


fig.add_subplot(1,3,1)
sns.countplot(df.gender, hue= df.SeniorCitizen)
plt.yticks(np.arange(0,2501,250))
plt.title("BAR PLOT OF GENDERS AND SENIORITY")

fig.add_subplot(1,3,2)
plt.pie(seniority0.gender.value_counts(), labels= seniority0.gender.value_counts().index, autopct= '%1.2f%%', startangle= 90)
plt.title('JUNIOR GENDERS', fontsize= 13)


fig.add_subplot(1,3,3)
plt.pie(seniority1.gender.value_counts(), labels= seniority1.gender.value_counts().index, autopct= '%1.2f%%', startangle= 0)
plt.title('SENIOR GENDERS', fontsize= 13)


plt.show()
```

```python
# Genders and Phone Service

yes_phone = df[df["PhoneService"]=="Yes"]
print("==GENDERS WHO HAVE PHONE SERVICE==")
print(yes_phone.gender.value_counts())

no_phone = df[df["PhoneService"]=="No"]
print("\n==GENDERS WHO DON'T HAVE PHONE SERVICE==")
print(no_phone.gender.value_counts())


fig= plt.figure(figsize= (16,5))

fig.add_subplot(1,3,1)
sns.countplot(df.gender, hue= df.PhoneService)
plt.yticks(np.arange(0,3001,250))
plt.title("BAR PLOT OF GENDERS AND PHONE SERVICE")

fig.add_subplot(1,3,2)
plt.pie(yes_phone.gender.value_counts(), labels= yes_phone.gender.value_counts().index, autopct= '%1.2f%%', startangle= 90)
plt.title('GENDERS WITH PHONE SERVICE', fontsize= 13)


fig.add_subplot(1,3,3)
plt.pie(no_phone.gender.value_counts(), labels= no_phone.gender.value_counts().index, autopct= '%1.2f%%', startangle= 0)
plt.title('GENDERS WITHOUT PHONE SERVICE', fontsize= 13)


plt.show()
```

```python
# Genders and Multiple Lines

yes_ml = df[df["MultipleLines"]=="Yes"]
print("==GENDERS WHO HAVE MULTIPLE LINES==")
print(yes_ml.gender.value_counts())

no_ml = df[df["MultipleLines"]=="No"]
print("\n==GENDERS WHO DON'T HAVE MULTIPLE LINES==")
print(no_ml.gender.value_counts())

no_phon_ser = df[df["MultipleLines"]=="No phone service"]
print("\n==GENDERS WHO DON'T HAVE MULTIPLE LINES==")
print(no_phon_ser.gender.value_counts())

fig= plt.figure(figsize= (12,12))

fig.add_subplot(2,2,1)
sns.countplot(df.gender, hue= df.MultipleLines)
plt.yticks(np.arange(0,3001,250))
plt.title("BAR PLOT OF GENDERS AND MULTIPLE LINES")

fig.add_subplot(2,2,2)
plt.pie(yes_ml.gender.value_counts(), labels= yes_ml.gender.value_counts().index, autopct= '%1.2f%%', startangle= 90)
plt.title('GENDERS WITH MULTIPLE LINES', fontsize= 13)


fig.add_subplot(2,2,3)
plt.pie(no_ml.gender.value_counts(), labels= no_ml.gender.value_counts().index, autopct= '%1.2f%%', startangle= 0)
plt.title('GENDERS WITHOUT MULTIPLE LINES', fontsize= 13)


fig.add_subplot(2,2,4)
plt.pie(no_phon_ser.gender.value_counts(), labels= no_phon_ser.gender.value_counts().index, autopct= '%1.2f%%', startangle= 0)
plt.title('GENDERS WITHOUT PHONE SERVICE', fontsize= 13)

plt.show()
```

```python
# Genders and Internet Service

intser_fib = df[df["InternetService"]=="Fiber optic"]
print("==GENDERS WHO HAVE FIBER OPTICS==")
print(intser_fib.gender.value_counts())

intser_dsl = df[df["InternetService"]=="DSL"]
print("\n==GENDERS WHO HAVE DSL==")
print(intser_dsl.gender.value_counts())

no_intser = df[df["InternetService"]=="No"]
print("\n==GENDERS WHO DON'T HAVE INTERNET SERVICE==")
print(no_intser.gender.value_counts())

fig= plt.figure(figsize= (12,12))

fig.add_subplot(2,2,1)
sns.countplot(df.gender, hue= df.InternetService)
plt.yticks(np.arange(0,3001,250))
plt.title("BAR PLOT OF GENDERS AND INTERNET SERVICE")

fig.add_subplot(2,2,2)
plt.pie(intser_fib.gender.value_counts(), labels= intser_fib.gender.value_counts().index, autopct= '%1.2f%%', startangle= 90)
plt.title('GENDERS WITH FIBER OPTICS', fontsize= 13)


fig.add_subplot(2,2,3)
plt.pie(intser_dsl.gender.value_counts(), labels= intser_dsl.gender.value_counts().index, autopct= '%1.2f%%', startangle= 0)
plt.title('GENDERS WITH DSL', fontsize= 13)


fig.add_subplot(2,2,4)
plt.pie(no_intser.gender.value_counts(), labels= no_intser.gender.value_counts().index, autopct= '%1.2f%%', startangle= 0)
plt.title('GENDERS WITHOUT INTERNET SERVICE', fontsize= 13)

plt.show()
```
