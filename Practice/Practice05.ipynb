```python
# Apartments and Month of Sale

aprts = product[product["Type of property"]=="Apartment"]
aprts.head()
```

```python
aprts["Type of property"].isnull().any()
```

```python
month_aparts = aprts["Month of sale"].value_counts().sort_index()
len(aprts["Month of sale"])
```

```python
aprts["Month of sale"].dropna(inplace=True)
```

```python
# No. Of Apartments Sold In Months

month_aparts.plot(kind="bar", color="navy", title="NO. OF APARTMENTS SOLD IN MONTHS",figsize=(9,6))
plt.xticks(np.arange(0,12), ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])
plt.xlabel("Months")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```
