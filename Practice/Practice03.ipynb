Prime Number

```python
num = int(input("Enter a number: "))

for i in range(2,num-1):
    if num % i == 0:
        print("<It Is Not A Prime Number>")
        break
else:
    print("<It Is A Prime Number>")
```

Fibbonaci Series

```python
def fib(n):
    a = 0
    b = 1
    if n <= 0:
        print("No Fibbonaci Series")
    elif n == 1:
        print(a)
    else:
        print(a)
        print(b)
        for i in range(2,n):
            c = a + b
            a = b
            b = c
            print(c)

fib(7)
```

Factorial

```python
def fact(x):
    f = 1
    for i in range(1,x+1):
        f = f * i
    print("The factorial of",x,"is",f)

num = int(input("Enter a number to return its factorial: "))
fact(num)
```

Counting Odd And Even

```python
def count(lst):
    even = 0
    odd = 0
    
    for i in lst:
        if i%2 == 0:
            even += 1
        else:
            odd += 1
    print(even,"even numbers and",odd,"odd numbers in your list")

list1 = [10,34,21,3,18,11,13,15,20]
count(list1)
```

Factorial with Recursion

```python
def fact(n):
    if n == 0:
        return 1
    
    return n * fact(n-1)

result = fact(5)
print(result)
```

Lambda Functions

```python
#Square of a number
f = lambda a: a*a
result = f(4)
print(result)
```

```python
#Adding Two Numbers
f = lambda a,b: a+b
result = f(2,5)
print(result)
```

Filter - Map - Reduce

```python
#Filtering even and odd numbers from list using filter()
nums = [3,4,6,2,3,8,10,11,98]

evens = list(filter(lambda a: a%2==0, nums))
odds = list(filter(lambda b: b%2!=0, nums))

print(evens)
print(odds)
```

```python
#Modifying Values through map()
num = [1,2,3,4,5,6]
doubles = list(map(lambda n: n*2, num))

print(doubles)
```

```python
#Sum of list values through reduce()
from functools import reduce
n = [2,3,4,1,5,6,10]
s = 0
print("== Through Loop ==")
for i in n:
    s+=i
print("Sum of list is",s)

print("\n== Through Reduce Function ==")
sum = reduce(lambda a,b: a+b, n)
print("Sum of list is",sum)
```
