```python
from bs4 import BeautifulSoup as bs
from bs4 import SoupStrainer as ss
import requests
```

```python
my_url = "https://www.aallyandsons.com/"
```

```python
response = requests.get(my_url)
```

```python
print(response)
```

```python
pg_html = response.text
```

```python
type(pg_html)
```

```python
soup = bs(pg_html, "html.parser")
type(soup)
```

```python
soup.div
```

```python
# find()
# find_all()
```

```python
anchor = soup.find("a")
anchor
```

```python
all_a = soup.find_all("a")
all_a
```

```python
# h4 with class card-title

h4_ct = soup.find_all("h4",{"class":"card-title"})
h4_ct
```

```python
len(h4_ct)
```

```python
# span with prices

span_price = soup.find_all("span",{"class":"price price--withTax price--main"})
span_price
```

```python
# a tags in h4-card-title 
for a in h4_ct:
    print(a.a)
```

```python
# hrefs from h4-card-title
for a in h4_ct:
    x = a.a
    print(x["href"])
    print(x.text)
```

```python
# Soup Strainer
strainer = ss("a")
soup = bs(pg_html, parse_only=strainer)
print(soup)
```

```python
# list of links

links = []
for a in h4_ct:
    links.append(a.a["href"])
    
print(links)
```

```python
# printing html codes of all the links

for i in links:
    html_str = requests.get(i).text
print(html_str)
```

```python
# Using Strainer and scraping the required data dirctly

for i in links:
    html_str = requests.get(i).text
    strainer = ss("h1",{"class":"productView-title"})
    soup = bs(html_str, parse_only=strainer)
    print(soup)
```

```python
for i in links:
    html_str = requests.get(i).text
    strainer = ss("span",{"class":"price price--withTax price--main"})
    soup = bs(html_str, parse_only=strainer)
    print(soup)
```
