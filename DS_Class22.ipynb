```python
# Extracting Total Females Survived
females_survived = df[(df['Sex']=='female') & (df['Survived']==1)]['Name'].values
```

```python
# Names of Females Survived
females_survived 
```

```python
#Result Males Survived
print("Males that survived from the Titanic incident are",len(males_survived),"in total")
```
