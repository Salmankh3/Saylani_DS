```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Task-01

```python
data = {'Investor_A': [96,181,88],
        'Investor_B':  [185,3,152],
        'Investor_C':   [39,29,142]
}
```

```python
df = pd.DataFrame(data, index=['Stocks', 'Bond', 'Real_Estate'])
df
```

```python
df.index.name = 'Type Of Investment'
df.columns.name = 'Investors'
df
```

```python
df[0:1].values.sum()
```

```python
df["Total"] = np.array([df[0:1].values.sum(),df[1:2].values.sum(),df[2:3].values.sum()])
df
```

```python
df["Investor_A"].sum()
```

```python
new_row = pd.Series(data={'Investor_A':df["Investor_A"].sum(), 'Investor_B':df["Investor_B"].sum(), 'Investor_C':df["Investor_C"].sum(), 'Total':df["Total"].sum()}, name='Total')
df = df.append(new_row,ignore_index=False)
df
```

```python
df2 = df[0:3]
df2
```

```python
df = df2.T
df
```

```python
df = df[0:3]
df
```

```python
width = 0.2
bar1 = np.arange(3)
bar2 = [i+width for i in bar1]
bar3 = [i+width for i in bar2]

plt.figure(figsize=(8,5))
plt.bar(bar1,df.Stocks,width,label="Stocks", color="k")
plt.bar(bar2,df.Bond,width,label="Bond",color="r")
plt.bar(bar3,df.Real_Estate,width,label="Real Estate",color="g")
plt.xticks(bar1+width,["Investor A", "Investor B", "Investor C"])
plt.yticks(np.arange(0,201,20))
plt.title("SIDE BY SIDE BAR PLOT", fontsize=14)
plt.legend()
plt.show()
```
