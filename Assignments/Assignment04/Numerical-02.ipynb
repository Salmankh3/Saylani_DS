```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Task-01

```python
list1 = []
flag = True
while flag:
    num = input("Enter a number: ")
    if num == "quit" or num == "Quit":
        flag = False
    else:
        num = int(num)
        list1.append(num)
list1 = sorted(list1)
```

```python
list1
```

```python
nums = pd.DataFrame({"Numbers":list1})
nums
```

```python
plt.figure(figsize=(10,6))
plt.plot(nums,"D-.k")
plt.title("PLOTTING NUMERICAL DATA", fontsize=15)
plt.xlabel("Index",fontsize=14)
plt.ylabel("Value",fontsize=14)
plt.xticks(np.arange(20))
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(nums.index,nums["Numbers"], width=0.7, color="green", alpha=0.8)
plt.title("VERTICAL BAR PLOT OF NUMERICAL DATA", fontsize=15)
plt.xlabel("Index",fontsize=14)
plt.ylabel("Value",fontsize=14)
plt.xticks(np.arange(20))
for index,data in enumerate(nums["Numbers"]):
    plt.text(x=index , y =data+1 , s=f"{data}" , fontdict=dict(fontsize=8))
plt.show()
```

```python
plt.figure(figsize=(8,10))
plt.barh(nums.index,nums["Numbers"], height=0.5, color="maroon", alpha=0.8)
plt.title("HORIZONTAL BAR PLOT OF NUMERICAL DATA", fontsize=15)
plt.xlabel("Value",fontsize=14)
plt.ylabel("Index",fontsize=14)
plt.yticks(np.arange(20))
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.title("PIE PLOT OF NUMERICAL DATA", fontsize=15)
plt.pie(nums, labels=nums.index)
plt.show()
```

# Task-02

```python
data = pd.DataFrame({
    "Interval Start":[8,55,101,147,193,239],
    "Interval End":[54,100,146,192,238,284],
    "Frequency":[nums[(nums["Numbers"]>=8) & (nums["Numbers"]<=54)].values.size,nums[(nums["Numbers"]>=55) & (nums["Numbers"]<=100)].values.size,nums[(nums["Numbers"]>=101) & (nums["Numbers"]<=146)].values.size,nums[(nums["Numbers"]>=147) & (nums["Numbers"]<=192)].values.size,nums[(nums["Numbers"]>=193) & (nums["Numbers"]<=238)].values.size,nums[(nums["Numbers"]>=239) & (nums["Numbers"]<=284)].values.size]
})

data
```

```python
tot_freq = data["Frequency"].sum()
data["Rel_Freq"] = (data["Frequency"]/tot_freq)
data["Rel_Freq_Percent"] = (data["Frequency"]/tot_freq)*100
```

```python
data
```

```python
plt.figure(figsize=(8,5))
plt.bar(data.index,data.Frequency, color=["blue","black","red","orange","green","gray"])
plt.xticks(np.arange(6),np.array(["8-54","55-100","101-146","147-192","193-238","239-284"]))
plt.title("BAR PLOT OF FREQUENCY DISTRIBUTION TABLE", fontsize=15)
plt.xlabel("Range",fontsize=14)
plt.ylabel("Frequency",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(data.index,data.Rel_Freq,"*--r")
plt.xticks(np.arange(6),np.array(["8-54","55-100","101-146","147-192","193-238","239-284"]))
plt.yticks(np.array([0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40,0.45]),np.array([0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40,0.45]))
plt.title("LINE PLOT OF RELATIVE FREQUENCY", fontsize=15)
plt.xlabel("Range",fontsize=14)
plt.ylabel("Relative Frequency",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(data.Rel_Freq_Percent,autopct="%1.2f%%",colors=["blue","yellow","red","green","orange","magenta"])
plt.title("PIE PLOT OF RELATIVE FREQUENCY PERCENTAGE", fontsize=15)
plt.legend(labels=np.array(["8-54","55-100","101-146","147-192","193-238","239-284"]))
```
