```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Task-01

```python
nums = pd.DataFrame({"Numbers":[1,9,22,24,32,33,41,44,57,66,70,73,75,76,79,82,87,89,95,100]})
```

```python
nums
```

```python
plt.figure(figsize=(10,6))
plt.plot(nums,"o-g")
plt.title("PLOTTING NUMERICAL DATA", fontsize=15)
plt.xlabel("Index",fontsize=14)
plt.ylabel("Value",fontsize=14)
plt.xticks(np.arange(20))
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(nums.index,nums["Numbers"], width=0.5, color="maroon", alpha=0.8)
plt.title("VERTICAL BAR PLOT OF NUMERICAL DATA", fontsize=15)
plt.xlabel("Index",fontsize=14)
plt.ylabel("Value",fontsize=14)
plt.xticks(np.arange(20))
plt.yticks(np.arange(0,101,10))
plt.show()
```

```python
plt.figure(figsize=(8,10))
plt.barh(nums.index,nums["Numbers"], height=0.5, color="orange", alpha=0.8)
plt.title("HORIZONTAL BAR PLOT OF NUMERICAL DATA", fontsize=15)
plt.xlabel("Value",fontsize=14)
plt.ylabel("Index",fontsize=14)
plt.yticks(np.arange(20))
plt.xticks(np.arange(0,101,10))
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.title("PIE PLOT OF NUMERICAL DATA", fontsize=15)
plt.pie(nums, labels=nums.index)
plt.show()
```

# Task-02

```python
data = pd.DataFrame({
    "Interval Start":[1,21,41,61,81],
    "Interval End":[20,40,60,80,100],
    "Frequency":[nums[(nums["Numbers"]>=1) & (nums["Numbers"]<=20)].values.size,nums[(nums["Numbers"]>=21) & (nums["Numbers"]<=40)].values.size,nums[(nums["Numbers"]>=41) & (nums["Numbers"]<=60)].values.size,nums[(nums["Numbers"]>=61) & (nums["Numbers"]<=80)].values.size,nums[(nums["Numbers"]>=81) & (nums["Numbers"]<=100)].values.size]
})

data
```

```python
tot_freq = data["Frequency"].sum()
data["Rel_Freq"] = (data["Frequency"]/tot_freq)
data["Rel_Freq_Percent"] = (data["Frequency"]/tot_freq)*100
```

```python
data
```

```python
plt.figure(figsize=(8,5))
plt.bar(data.index,data.Frequency, color=["black","red","orange","green","maroon"])
plt.xticks(np.arange(5),np.array(["1-20","21-40","41-60","61-80","81-100"]))
plt.title("BAR PLOT OF FREQUENCY DISTRIBUTION TABLE", fontsize=15)
plt.xlabel("Range",fontsize=14)
plt.ylabel("Frequency",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(data.index,data.Rel_Freq,"o-k")
plt.xticks(np.arange(5),np.array(["1-20","21-40","41-60","61-80","81-100"]))
plt.yticks(np.array([0.05,0.10,0.15,0.20,0.25,0.30,0.35]),np.array([0.05,0.10,0.15,0.20,0.25,0.30,0.35]))
plt.title("LINE PLOT OF RELATIVE FREQUENCY", fontsize=15)
plt.xlabel("Range",fontsize=14)
plt.ylabel("Relative Frequency",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(data.Rel_Freq_Percent,autopct="%1.2f%%")
plt.title("PIE PLOT OF RELATIVE FREQUENCY PERCENTAGE", fontsize=15)
plt.legend(labels=np.array(["1-20","21-40","41-60","61-80","81-100"]))
```
