```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Task-01

```python
nums = pd.DataFrame({"Numbers":[1,9,22,24,32,33,41,44,57,66,70,73,75,76,79,82,87,89,95,100]})
```

```python
nums
```

```python
plt.figure(figsize=(10,6))
plt.plot(nums,"o-g")
plt.title("PLOTTING NUMERICAL DATA", fontsize=15)
plt.xlabel("Index",fontsize=14)
plt.ylabel("Value",fontsize=14)
plt.xticks(np.arange(20))
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(nums.index,nums["Numbers"], width=0.5, color="maroon", alpha=0.8)
plt.title("VERTICAL BAR PLOT OF NUMERICAL DATA", fontsize=15)
plt.xlabel("Index",fontsize=14)
plt.ylabel("Value",fontsize=14)
plt.xticks(np.arange(20))
plt.yticks(np.arange(0,101,10))
plt.show()
```

```python
plt.figure(figsize=(8,10))
plt.barh(nums.index,nums["Numbers"], height=0.5, color="orange", alpha=0.8)
plt.title("HORIZONTAL BAR PLOT OF NUMERICAL DATA", fontsize=15)
plt.xlabel("Value",fontsize=14)
plt.ylabel("Index",fontsize=14)
plt.yticks(np.arange(20))
plt.xticks(np.arange(0,101,10))
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.title("PIE PLOT OF NUMERICAL DATA", fontsize=15)
plt.pie(nums, labels=nums.index)
plt.show()
```
