```python
std_rec = {"Name":["Salman","Uzaif","Fayyaz","Nabeel","Umair"], 
           "Age":[21,23,24,25,20], 
           "Python":[75,78,87,72,90], 
           "Numpy":[77,83,88,np.NaN,73], 
           "Pandas":[95,92,78,82,80]}

data = pd.DataFrame(std_rec)
```

```python
new_rec = {"Name":"Talha", 
           "Age":21, 
           "Python":40, 
           "Numpy":80, 
           "Pandas":70}

data = data.append(new_rec, ignore_index=True)
```

```python
data["Total_Obt"] = data["Python"] + data["Numpy"] + data["Pandas"]
```

```python
data
```

```python
data["Result"]=["Pass" if i >= 150 else "Fail" for i in data["Total_Obt"]]
```

```python
data
```

```python
data["Actual_Result"] = ["Pass" if (py>=50 and num>=50 and pan>=50) else "Fail" for py,num,pan in zip(data["Python"],data["Numpy"],data["Pandas"])]
```

```python
data
```

```python
data["Percentage"] = round((data["Total_Obt"] / 300)*100,2)
```

```python
data
```

```python
data["Grade"] = ["A" if g>=80 else "B" if g>=70 else "C" if g>=60 else "D" if g>=50 else "F" for g in data["Percentage"]]
```

```python
data
```

```python
# Dropping from our data
data.drop(index=5, inplace=True) # Row dropped
```
