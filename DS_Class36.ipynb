<b>Task 03</b>

Performing Some Data Analysis

```python
customer["Gender"].value_counts()
```

```python
# Pie Chart Of Genders

plt.figure(figsize=(5,5))
plt.title("PIE CHART OF GENDERS")
plt.pie(customer["Gender"].value_counts(),colors=["orange","green","red"],autopct="%1.2f%%",labels=["Males","Females","Firms"])
plt.show()
```

```python
# Bar Plot of Gender Frequencies

customer["Gender"].value_counts().plot(kind="bar",title="BAR PLOT OF GENDERS")
```

------------------------

```python
customer["Deal satisfaction"].value_counts()
```

```python
# Pie Chart Of Deal Satisfaction

plt.figure(figsize=(5,5))
plt.title("PIE CHART OF DEAL SATISFACTION")
plt.pie(customer["Deal satisfaction"].value_counts(),colors=["orange","green","red","blue","aqua"],autopct="%1.2f%%",labels=["5 Ratings","4 Ratings","3 Ratings","2 Ratings","1 Ratings"])
plt.show()
```

```python
# Bar Chart Of Deal Satisfaction Frequencies

customer["Deal satisfaction"].value_counts().plot(kind="bar",title="BAR PLOT OF DEAL SATISFACTION")
plt.xticks(np.arange(0,5),["5 Ratings","4 Ratings","3 Ratings","2 Ratings","1 Ratings"])
plt.figure(figsize=(20,5))
plt.show()
```

------------------

```python
females = customer[customer["Gender"]=="F"]
females["Deal satisfaction"].value_counts()
```

```python
# Pie Chart Of No.Of Ratings Given by Females

plt.figure(figsize=(5,5))
plt.title("PIE CHART OF DEAL SATISFACTION (FEMALES)")
plt.pie(females["Deal satisfaction"].value_counts(),colors=["orange","green","red","blue","aqua"],autopct="%1.2f%%",labels=["5 Ratings","4 Ratings","3 Ratings","2 Ratings","1 Ratings"])
plt.show()
```

```python
males = customer[customer["Gender"]=="M"]
males["Deal satisfaction"].value_counts()
```

```python
# Pie Chart Of No.Of Ratings Given by Males

plt.figure(figsize=(5,5))
plt.title("PIE CHART OF DEAL SATISFACTION (FEMALES)")
plt.pie(males["Deal satisfaction"].value_counts(),colors=["orange","green","red","blue","aqua"],autopct="%1.2f%%",labels=["5 Ratings","4 Ratings","3 Ratings","2 Ratings","1 Ratings"])
plt.show()
```

```python
firms = customer[customer["Gender"]=="Firm"]
firms["Deal satisfaction"].value_counts()
```

```python
# Pie Chart Of No.Of Ratings Given by Firms

plt.figure(figsize=(5,5))
plt.title("PIE CHART OF DEAL SATISFACTION (FIRMS)")
plt.pie(firms["Deal satisfaction"].value_counts(),colors=["lightgreen","red","aqua","orange"],autopct="%1.2f%%",labels=["5 Ratings","4 Ratings","3 Ratings","1 Ratings"])
plt.show()
```
