```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import sem, t
from scipy import mean
```

```python
confidence = 0.98

n = len(data)
m = mean(data)
std_err = sem(data)
h = std_err * t.ppf((1 + confidence) / 2, n - 1)

start = m - h
print(start)

end = m + h
print(end)
```

```python
print("At Confidence of ",confidence,", the range in which mean lies is from ",start.Salary.round(2),"to ",end.Salary.round(2))
```
